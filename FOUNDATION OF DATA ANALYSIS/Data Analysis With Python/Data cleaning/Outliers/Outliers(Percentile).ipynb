```python
min = df["scholarship_exam_marks"].quantile(0.02)
max = df["scholarship_exam_marks"].quantile(0.98)
```

```python
print(min, max)
```

```python
df[(df["scholarship_exam_marks"]< min )| (df["scholarship_exam_marks"]> max)]
```

## Trimming

```python
trim_df = df[(df["scholarship_exam_marks"]> min )& (df["scholarship_exam_marks"]< max)]
```

```python
sns.boxplot(trim_df["scholarship_exam_marks"])
```

## Capping


```python
df["scholarship_exam_marks"] = np.where(df["scholarship_exam_marks"] < min , min, np.where(df["scholarship_exam_marks"]> max,
                                       max, df["scholarship_exam_marks"]))
```

```python
new_df = df.copy()
new_df["scholarship_exam_marks"] = np.where(df["scholarship_exam_marks"] < min , min, np.where(df["scholarship_exam_marks"]> max,
                                       max, df["scholarship_exam_marks"]))
```
