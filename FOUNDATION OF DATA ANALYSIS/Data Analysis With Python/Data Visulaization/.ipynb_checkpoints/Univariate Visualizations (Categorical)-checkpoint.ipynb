```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.columns # .HEAD()
```

```python
df.nunique()
```

```python
sns.countplot(x = df["room_type"])
```
