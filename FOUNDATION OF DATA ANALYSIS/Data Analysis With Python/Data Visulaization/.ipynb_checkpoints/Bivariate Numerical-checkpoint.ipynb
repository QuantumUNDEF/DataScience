```python
# importing library
import pandas as pd
import seaborn as sns
```

```python
#working on inbuilt dataset of seaborn library
df = sns.load_dataset("titanic")
```

```python
#see how data looks like
df.head()
```

 Fisrt we will analyise the AGE and CLASS column

```python
#ploting the barplot of age and class 
sns.barplot(data = df, x = "class", y = "age")#mean value default
```

```python
#bar plot is use to plot average value of categorical data
df[df["class"] == "First"]["age"]
```

```python
#changing the estimator to the median from the 
sns.barplot(data = df, x = "class", y = "age", estimator = "median")
```

```python
# adding the hue parameter
sns.barplot(data = df, x = "class", y = "age", hue = "sex")
```

```python
sns.barplot(x ="class", y= "age",hue = "sex",estimator="sum", data = df)
```

```python
sns.boxplot(x = "sex", y= "age", data = df)
```

```python
sns.boxplot(x = "sex", y= "age", hue = "class", data = df)
```

```python
sns.distplot(df[df["survived"]==0]["age"], hist = False)
sns.distplot(df[df["survived"]==1]["age"], hist = False)
```

```python
sns.violinplot(x = "sex", y = "fare", data = df)
```
